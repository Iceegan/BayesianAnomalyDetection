# Plotting Some of the Data

```
import matplotlib
matplotlib.rcParams.update({'font.size': 22,
                            'legend.fontsize':12})

fig, ax = plt.subplots(2, 2, figsize=(24,12))
event1Bus.filter(regex='B.VA_Mag').plot(ax=ax[0,0])
ax[0,0].set_title('Event 1')
ax[0,0].set_ylabel('Voltage Magnitude (p.u.)')
ax[0,0].set_xlabel('Time (s)')
event2Bus.filter(regex='B.VA_Mag').plot(ax=ax[1,0])
ax[0,1].set_title('Event 2')
ax[0,1].set_ylabel('Voltage Magnitude (p.u.)')
ax[0,1].set_xlabel('Time (s)')
event3Bus.filter(regex='B.VA_Mag').plot(ax=ax[0,1])
ax[1,0].set_title('Event 3')
ax[1,0].set_ylabel('Voltage Magnitude (p.u.)')
ax[1,0].set_xlabel('Time (s)')
event4Bus.filter(regex='B.VA_Mag').plot(ax=ax[1,1])
ax[1,1].set_title('Event 4')
ax[1,1].set_ylabel('Voltage Magnitude (p.u.)')
ax[1,1].set_xlabel('Time (s)')
plt.tight_layout()
plt.show()
```
